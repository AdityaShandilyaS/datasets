```python
import numpy as np

def split_train_test(data, test_ratio):
    seed = np.random.seed(42)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def data_preparation_pipeline(dataset, type=''):
    housing_numeric = dataset.loc[:,dataset.dtypes != 'object']
    housing_objects = dataset.loc[:,dataset.dtypes == 'object']
    
    housing_numeric = housing_numeric.drop("Id", axis=1)
    
    housing_objects = housing_objects.astype(str)
    
    housing_objects["Alley"].fillna(value="noAlley", inplace=True)
    housing_objects["MasVnrType"].fillna(value="noVeneer", inplace=True)
    housing_objects["Electrical"].fillna(value=housing_objects["Electrical"].median, inplace=True)
    basement_fields = ("BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2")
    for x in basement_fields:
        housing_objects[x].fillna(value="noBsmt", inplace=True)
    housing_objects["FireplaceQu"].fillna(value="noFire", inplace=True)
    garage_fields = ("GarageType", "GarageFinish", "GarageQual", "GarageCond")
    for i in garage_fields:
        housing_objects[i].fillna(value="noGrge", inplace=True)
    housing_objects["PoolQC"].fillna(value="noPool", inplace=True)
    housing_objects["Fence"].fillna(value="noFnce", inplace=True)
    housing_objects["MiscFeature"].fillna(value="noMscFt", inplace=True)
    
    housing_objects_encoded = housing_objects.apply(LabelEncoder().fit_transform)
    
    housing_final = pd.concat([housing_objects_encoded, housing_numeric], axis=1)
    housing_final.fillna(value=0, inplace=True)
       
    if (type != 'test'):
        housing_final_labels = housing_final["SalePrice"]
        housing_final.drop(["SalePrice"], axis=1, inplace=True)
    
    numpy_arrays = housing_final.to_numpy()
    prepared_dataset = StandardScaler().fit_transform(numpy_arrays)
#     prepared_dataset = MinMaxScaler().fit_transform(numpy_arrays)
    
    if (type != 'test'):
        return prepared_dataset, housing_final_labels
    
    return prepared_dataset
```

```python
def split_train_labels(dataset):
    train_labels = dataset["SalePrice"]
    dataset.drop("SalePrice", axis=1, inplace=True)
    return train_labels

def split_train_val(dataset):
    train_housing, val_housing = split_train_test(dataset, 0.2)
    train_housing_labels = train_housing["SalePrice"]
    train_housing = train_housing.drop("SalePrice", axis=1)
    val_housing_labels = val_housing["SalePrice"]
    val_housing = val_housing.drop("SalePrice", axis=1)
    return train_housing, train_housing_labels, val_housing, val_housing_labels
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, dataset, labels):
    train_set, val_set, train_labels, val_labels = train_test_split(dataset, labels, test_size=0.2)
    train_errors, val_errors = [], []
    
    for m in range(1, len(train_set)):
        model.fit(train_set[:m], train_labels[:m])
        train_predict = model.predict(train_set[:m])
        val_predict = model.predict(val_set)
        train_errors.append(mean_squared_error(train_predict, train_labels[:m]))
        val_errors.append(mean_squared_error(val_predict, val_labels))
        
    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "g", linewidth=3, label="val")
```
